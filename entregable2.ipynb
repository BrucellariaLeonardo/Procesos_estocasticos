```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

# Parametros de simulacion

```python
#Parametros de simulacion
T = 4 #rango de la uniforme
realizaciones = 100 # numero de muestras
muestras = 100 # cantidad de pasos de tiempo que 
```

```python
# vectores de tiempo para la simulacion
T1 = np.linspace(-1,T+1, muestras)
T2 = np.linspace(-1,T+1, muestras)
```

# Simulacion

```python
#genero el proceso
X= np.random.uniform(0,T, size = (realizaciones, muestras)) # genero muestras de la VA c desde t = 0 hasta t = T para cada realiacion
for i in range(muestras): #le resto el valor del tiempo de la muestra a los valores de la VA
    X[:][i] = T1[i] - X[:][i] 
X1 = np.heaviside(X,0)   #aplico el escalon
```

```python
matR = X1 @ X1.T / muestras #calculo de la matrix de correlacion
```

```python
plt.imshow(matR, extent=[T1[0],T1[-1],T2[0],T2[-1]], origin='lower', aspect='auto', cmap = "plasma")
plt.colorbar()

plt.xlabel("t1")
plt.ylabel("t2")
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
x =  np.linspace(0,muestras,muestras)
y =  np.linspace(0,muestras,muestras)
X,Y = np.meshgrid(x,y)
im = ax.plot_surface(X,Y, matR, cmap = 'plasma')
ax.contourf(X,Y,matR,zdir = 'z', offset = -0.3, cmap = 'plasma') #sombra del grafico
plt.gca().azim = -45 #rotacion camara
plt.gca().elec = 10 #altura de la camara
fig.colorbar(im, shrink = 0.8, pad = 0.1) #configuracion indicador barrar de colores

plt.show()
```
